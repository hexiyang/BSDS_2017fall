```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
# Read Data
get_10 = pd.read_table('10-100/Get.dat', sep='::', header=0, index_col=False, engine='python')
post_10 = pd.read_table('10-100/Post.dat', sep='::', header=0, index_col=False, engine='python')
get_100 = pd.read_table('100-100/Get.dat', sep='::', header=0, index_col=False, engine='python')
post_100 = pd.read_table('100-100/Post.dat', sep='::', header=0, index_col=False, engine='python')
```

```python
get_10[:5]
```

```python
# Generate Figures
get10fig = get_10.plot(title='Latency for Get with 10 Threads and 100 Iterations', 
                       x='Timestamp', y='Latency', kind='line', figsize=(15,5))
post10fig = post_10.plot(title='Latency for Post with 10 Threads and 100 Iterations', 
                       x='Timestamp', y='Latency', kind='line', figsize=(15,5))
get100fig = get_100.plot(title='Latency for Get with 100 Threads and 100 Iterations', 
                       x='Timestamp', y='Latency', kind='line', figsize=(150,5))
post100fig = post_100.plot(title='Latency for Post with 100 Threads and 100 Iterations', 
                       x='Timestamp', y='Latency', kind='line', figsize=(150,5))
```

```python
# Save the figures
get10fig.get_figure().savefig('get_10_100.jpeg');
post10fig.get_figure().savefig('post_10_100.jpeg');
get100fig.get_figure().savefig('get_100_100.jpeg');
post100fig.get_figure().savefig('post_100_100.jpeg');
```
